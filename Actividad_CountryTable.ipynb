**En esta parte se aplica un filtro de busqueda con conectores bsicos como es el caso del conector &, or y not.**

Conector &

```python
#Filtro Y
filtro7= df[(df["life_expectancy"] <=70) & (df["region"]== 'South America')]
filtro7
```

```python
#Filtro Y
filtro7b= df[(df["independence_year"] > 1910) & (df["region"]== "Caribbean")]
filtro7b
```

Conector O

```python
#Filtro O
filtro8= df[(df["life_expectancy"]>80) | (df["region"]== "Caribbean")]
filtro8
```

```python
#Filtro O
filtro8b= df[(df["population"]>50000000) | (df["continent"]== "Asia")]
filtro8b
```

Conector NOT

```python
#Filtro not
filtro9=df[~(df["surface_area"] == 4000000) & ~(df["government_form"]== "Republic")]
filtro9
```

```python
#Filtro not
filtro9b=df[~(df["continent"] == 'Europe') & ~(df["government_form"]== "Republic")]
filtro9b
```
